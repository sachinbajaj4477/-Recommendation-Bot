```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_Desktop_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Selling Price'][4135].split('₹')[1].replace(',','')
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'][0].split('+')[0]
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].unique()
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '50'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df['Product_Name'].unique()
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df
```

```python
import random

# Popular laptop brands
popular_laptop_brands = ["Dell", "HP", "Lenovo", "Apple", "Asus", "Acer"]

# Replace existing brands with random popular laptop brands
df['Brand'] = df['Brand'].apply(lambda x: random.choice(popular_laptop_brands) if x not in popular_laptop_brands else x)
```

```python
df
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

### Inferences

* Selling Price:
The distribution of selling prices shows a concentration around lower values, with a few products having significantly higher selling prices.
Most products seem to be priced on the lower end, possibly indicating a diverse range of products in terms of pricing.

* MRP:
The distribution of Maximum Retail Prices (MRP) is similar to the selling price distribution.
The majority of products have MRPs in the lower range, but there are outliers with much higher MRPs.

* Items Bought Last Month:
The distribution of items bought last month is right-skewed, suggesting that the majority of products have a lower number of items bought.
There are a few products that have a relatively higher number of items bought, indicating their popularity.

* Numeric Ratings:
The distribution of numeric ratings shows a concentration around 4.2, indicating that most products have positive ratings.
There is a spread in ratings, with some products having ratings below 4.0 and others above.

* Total Ratings:
The distribution of total ratings is right-skewed, indicating that most products have a lower total number of ratings.
Some products with higher total ratings may be more popular or have been on the market for a longer time.

* Offer Percentage:
The distribution of offer percentages is right-skewed, with a concentration of products having relatively high offer percentages.
Many products have offer percentages above 20%, suggesting the presence of discounts or promotional activities.

* General Inferences:
The dataset includes a diverse range of products with varying selling prices, MRPs, and offer percentages.
Popularity, as indicated by items bought last month and total ratings, varies among products.
The majority of products have positive numeric ratings, but there is variability in customer satisfaction.

* Selling Price vs. MRP:

* Items Bought Last Month vs. Numeric Ratings:
There is no clear linear relationship between the number of items bought last month and numeric ratings.
The popularity of a product, as measured by items bought, does not necessarily correlate strongly with higher ratings.

* Total Ratings vs. Numeric Ratings:
There is a positive linear relationship between total ratings and numeric ratings.
Products with higher total ratings tend to have higher numeric ratings, suggesting that well-received products attract more ratings.

* Offer Percentage vs. Selling Price:
There is a negative correlation between offer percentage and selling price.
Higher-priced products tend to have lower offer percentages, indicating that discounts are less common for expensive items.

* Offer Percentage vs. Numeric Ratings:
There is no clear linear relationship between offer percentage and numeric ratings.
The presence of discounts (offer percentage) does not strongly correlate with higher or lower numeric ratings.

* Brand vs. Numeric Ratings:
Different brands have varying distributions of numeric ratings.
Some brands have a higher median rating, indicating better overall customer satisfaction, while others have a wider range of ratings.

* General Inferences:
The alignment between selling prices and MRPs suggests consistency in pricing strategies.
Popularity, as measured by the number of items bought last month, does not always correlate with higher numeric ratings.
Higher total ratings tend to be associated with higher numeric ratings, indicating a positive relationship between overall popularity and customer satisfaction.
Higher-priced products generally have lower offer percentages, while the presence of discounts does not guarantee higher ratings.
Brands play a role in determining customer satisfaction, with some brands consistently receiving better ratings than others.
