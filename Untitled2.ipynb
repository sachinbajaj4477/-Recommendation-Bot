```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_DSLRCameras_final_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month'].unique()
```

```python
df['Items Bought Last Month'].value_counts()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '100'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df
```

```python
df['Brand'].unique()
```

```python
df['Brand'].value_counts()
```

```python
df['Brand'] = df['Brand'].replace('Eos', 'EOS')
```

```python
orgbrand=['EOS', 'Z50', 'M50', 'RP', 'D7500', 'Alpha', 'RX10',
                      'D850', 'X100V', 'K-3', 'OM-D', 'D7000', 'Canon', 'Lumix']
```

```python
def replace_with_nan(brand, orgbrand):
    if brand in orgbrand:
        return brand
    else:
        return np.nan

df['Brand'] = df['Brand'].apply(lambda x: replace_with_nan(x, orgbrand))
```

```python
df.isnull().sum()
```

```python
famous_camera_brands = ['Fujifilm', 'Nikon', 'Sony']
```

```python
import random

def replace_nan_with_random_brand(brand):
    if pd.isnull(brand):
        return random.choice(famous_camera_brands)
    else:
        return brand
    
# Apply the function to replace NaN values in the 'Brand' column
df['Brand'] = df['Brand'].apply(replace_nan_with_random_brand)
```

```python
df
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```
