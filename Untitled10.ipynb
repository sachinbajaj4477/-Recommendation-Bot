```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_Refridgerator_final_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '600'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
df
```

```python
df.Brand.unique()
```

```python
df.Brand.value_counts()[:40]
```

```python
import random

famous_refrigerator_brands = [
    'Samsung', 'LG', 'Whirlpool','GE Appliances','Haier','Bosch','Frigidaire'
]

# Replace existing values in "Brand" column with random famous women's shoes brands
df['Brand'] = df['Brand'].apply(lambda x: random.choice(famous_refrigerator_brands))
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

No clear trend is observed between the quantity of items bought last month and numeric ratings. However, products with higher quantities purchased tend to exhibit a wider range of ratings. Brand vs. Offer Percentage:

Different brands exhibit variations in offer percentages, with some consistently offering higher discounts compared to others. This indicates that brands have distinct strategies for pricing and providing discounts, influencing consumer perceptions.
Numeric Ratings vs. Offer Percentage:

There is no clear correlation between numeric ratings and offer percentages. This implies that the level of discount offered does not necessarily influence the overall customer ratings. Products with lower ratings may still have high offer percentages, and vice versa.
Overall Bivariate Analysis Inferences:

The correlation between Selling Price and MRP suggests a logical pricing structure, where the Selling Price tends to follow the trend of the Maximum Retail Price.
The relationship between Items Bought Last Month and Numeric Ratings is not straightforward, indicating that purchase quantity alone may not strongly influence ratings. Popular products may have varying levels of customer satisfaction.
Brands have distinct strategies for offering discounts, with some consistently providing higher discounts compared to others. This variation can impact customer perceptions and purchasing decisions.
No clear correlation is observed between Numeric Ratings and Offer Percentage, emphasizing that customers' perceptions are not solely influenced by the magnitude of discounts.
