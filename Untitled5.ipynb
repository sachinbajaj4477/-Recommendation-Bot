```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_Kitchen&HomeAppliances_final_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '50'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

```python
df.info()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].astype(int)
```

### Inferences

* Selling Price:
The distribution of selling prices is diverse, with a concentration around lower values.
There are products with a broad range of selling prices, including some high-priced items.

* MRP:
The distribution of Maximum Retail Prices (MRP) is right-skewed.
Most products have MRPs in the lower to mid-range, with a few products having higher MRPs.

* Items Bought Last Month:
The distribution of items bought last month indicates variability in product popularity.
Some products have lower demand (50 items), while others have higher demand (300 to 40,000 items).

* Numeric Ratings:
The distribution of numeric ratings is right-skewed, with a concentration around 4.0.
Most products have positive ratings, but there are a few with lower ratings.

* Total Ratings:
The distribution of total ratings is right-skewed, with a few products receiving a significantly higher number of ratings.
Most products have a moderate number of total ratings.

* Offer Percentage:
The distribution of offer percentages is right-skewed.
Many products have relatively high offer percentages, indicating the presence of discounts.

* General Inferences:
The dataset includes a mix of products with varied selling prices, MRPs, and popularity.
Numeric ratings are generally positive, indicating overall customer satisfaction.
Total ratings suggest that most products have a moderate number of ratings, but some products have attracted a large number of reviews.
Offer percentages vary, with some products having notable discounts.

### Inferences

* Selling Price vs. MRP:
Selling prices and MRPs show a strong positive linear relationship.
The increase in selling price corresponds to a proportional increase in MRP, indicating consistent pricing strategies.

* Items Bought Last Month vs. Numeric Ratings:
There is no clear linear relationship between the number of items bought last month and numeric ratings.
The popularity of a product, as measured by items bought, does not necessarily correlate strongly with higher ratings.

* Total Ratings vs. Numeric Ratings:
There is a positive linear relationship between total ratings and numeric ratings.
Products with higher total ratings tend to have higher numeric ratings, suggesting that well-received products attract more ratings.

* Offer Percentage vs. Selling Price:
There is no clear linear relationship between offer percentage and selling price.
Higher-priced products do not consistently have higher or lower offer percentages.

* Offer Percentage vs. Numeric Ratings:
There is no clear linear relationship between offer percentage and numeric ratings.
The presence of discounts (offer percentage) does not strongly correlate with higher or lower numeric ratings.

* Brand vs. Numeric Ratings:
Different brands have varying distributions of numeric ratings.
Some brands have a higher median rating, indicating better overall customer satisfaction, while others have a wider range of ratings.

* General Inferences:
The strong positive correlation between selling price and MRP suggests consistent pricing strategies across products.
Popularity, as measured by the number of items bought last month, does not strongly correlate with higher numeric ratings.
Higher total ratings tend to be associated with higher numeric ratings, indicating a positive relationship between overall popularity and customer satisfaction.
The presence of discounts (offer percentage) does not consistently correlate with higher or lower ratings.
Brands play a role in determining customer satisfaction, with some brands consistently receiving better ratings than others.
