```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_washingMachine_final_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '1K'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)

df
```

```python
df.Brand.unique()
```

```python
df.Brand.value_counts()[:40]
```

```python
df['Brand'] = df['Brand'].replace('Toshiba', 'TOSHIBA')
```

```python
orgbrand=['LG', 'Samsung', 'Whirlpool', 'Voltas', 'Godrej', 'IFB', 'Bosch',
                      'Panasonic', 'Lloyd', 'TOSHIBA', 'Siemens']
```

```python
def replace_with_nan(brand, orgbrand):
    if brand in orgbrand:
        return brand
    else:
        return np.nan

df['Brand'] = df['Brand'].apply(lambda x: replace_with_nan(x, orgbrand))
```

```python
df.isnull().sum()
```

```python
famous_wm_brands = ['Kenmore', 'Electrolux', 'GE Appliances','Maytag','Miele']
```

```python
import random

def replace_nan_with_random_brand(brand):
    if pd.isnull(brand):
        return random.choice(famous_wm_brands)
    else:
        return brand
    
# Apply the function to replace NaN values in the 'Brand' column
df['Brand'] = df['Brand'].apply(replace_nan_with_random_brand)
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

### Inferences

Selling Price:

The Selling Price of washing machines in the dataset varies widely, ranging from lower to higher values. This indicates a diverse pricing structure within different brands and models of washing machines.

MRP (Maximum Retail Price):

The Maximum Retail Price (MRP) of washing machines also shows significant variation across different models. Some machines have higher MRPs, possibly reflecting advanced features or brand positioning.

Items Bought Last Month:

The quantity of items bought last month for the washing machines in the dataset varies, suggesting differences in popularity among different models. Some machines have higher recent sales compared to others.

Numeric Ratings:

Numeric Ratings for the washing machines range from 4.1 to 4.5 out of 5 stars. This indicates a generally positive customer sentiment towards the products, with most products having good ratings.

Offer Percentage:

Offer percentages for washing machines range from around 20% to over 90%. This reflects diverse pricing strategies, with some machines having higher discounts, possibly to stimulate sales.

Overall Univariate Analysis Inferences:

Washing machines in the dataset have a diverse pricing structure, catering to different market segments.
Customers show a preference for washing machines with positive Numeric Ratings, indicating satisfaction with the products.
The dataset includes a variety of brands, reflecting diverse choices available in the market.
Differences in recent sales (Items Bought Last Month) indicate variations in the popularity of different washing machine models.
The wide range of Offer Percentages suggests distinct pricing and promotional strategies for different washing machine models.

### Inferences

Selling Price vs. MRP:

There is a positive correlation between Selling Price and MRP. As the Maximum Retail Price (MRP) increases, the Selling Price tends to increase. This suggests that washing machines with higher MRPs are generally sold at higher prices.

Items Bought Last Month vs. Numeric Ratings:

There is no clear correlation between the quantity of items bought last month and numeric ratings. However, products with higher quantities purchased tend to have a wider range of ratings. This could indicate that popular washing machines receive a diverse set of reviews, with both positive and negative feedback.

Brand vs. Offer Percentage:

Different brands exhibit variations in offer percentages. Some brands consistently offer higher discounts compared to others. This insight is valuable for both customers looking for discounts and brands aiming to optimize pricing strategies.

Numeric Ratings vs. Offer Percentage:

There is no clear correlation between numeric ratings and offer percentages. This suggests that the level of discount offered does not necessarily influence the overall customer ratings. Washing machines with lower ratings may still have high offer percentages.

Overall Bivariate Analysis Inferences:

Selling Price tends to follow the trend of Maximum Retail Price, indicating a logical pricing structure.
The relationship between Items Bought Last Month and Numeric Ratings is not straightforward, indicating that purchase quantity alone may not strongly influence ratings.
Brands have distinct strategies for offering discounts, with some consistently providing higher discounts.
No clear correlation is observed between Numeric Ratings and Offer Percentage, emphasizing that customers' perceptions are not solely influenced by discounts.
