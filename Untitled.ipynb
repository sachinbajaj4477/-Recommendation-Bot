```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Selling Price'][861].split('₹')[1].replace(',','')
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'][0].split('+')[0]
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df
```

```python
df['Items Bought Last Month'].unique()
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '100'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
df['MRP'][0].replace(',','')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
```

```python
df
```

```python
df['Brand'] = 'AC'
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

```python
import random
# Popular brand names
popular_brands = ["Samsung","LG","Haier","Voltas","Panasonic","Whirlpool"]

# Replace "AC" with random popular brand names
df['Brand'] = df['Brand'].apply(lambda x: random.choice(popular_brands) if x == "AC" else x)
```

### Inferences

* Selling Price Distribution:
The majority of products have a selling price in the range of Rs30,000 to Rs40,000.
There are a few products with significantly higher selling prices, potentially indicating premium or specialized items.

* MRP Distribution:
The distribution of Maximum Retail Price (MRP) is similar to the selling price.
The MRP of most products falls in the range of Rs50,000 to Rs70,000.

* Items Bought Last Month Distribution:
The distribution is right-skewed, indicating that most products have a lower number of items bought last month.
There are a few products that have a relatively higher number of items bought, suggesting their popularity

* Numeric Ratings Distribution:
The majority of products have numeric ratings around 4.0.
There is a spread in ratings, with some products having ratings below 4.0 and others above.

* Offer Percentage Distribution:
The distribution is right-skewed, with a concentration of products having high offer percentages.
Many products have offer percentages above 50%, indicating a significant discount or promotional activity.

* General Inferences:
Products with higher selling prices and MRPs tend to have positive numeric ratings.
The distribution of items bought last month suggests that most products have a moderate level of popularity, but there are a few highly popular items.
The majority of products have ratings around 4.0, indicating a generally positive sentiment among customers.
There is a wide range of offer percentages, with many products having substantial discounts.
