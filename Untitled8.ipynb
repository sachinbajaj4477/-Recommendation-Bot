```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_Monitor_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '50'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
```

```python
df
```

```python
df.Brand.unique()
```

```python
df.Brand.value_counts()[:30]
```

```python
orgbrand=['Dell','Samsung','LG','ASUS','Acer','BenQ','HP','ViewSonic','AOC','Philips']
```

```python
def replace_with_nan(brand, orgbrand):
    if brand in orgbrand:
        return brand
    else:
        return np.nan

df['Brand'] = df['Brand'].apply(lambda x: replace_with_nan(x, orgbrand))
```

```python
df.isnull().sum()
```

```python
famous_monitor_brands = ['LG','ViewSonic','Philips','Dell','BenQ','Acer','ASUS','AOC','HP','Samsung','Lenovo',
                       'MSI','Sony','NEC']
```

```python
import random

def replace_nan_with_random_brand(brand):
    if pd.isnull(brand):
        return random.choice(famous_monitor_brands)
    else:
        return brand
    
# Apply the function to replace NaN values in the 'Brand' column
df['Brand'] = df['Brand'].apply(replace_nan_with_random_brand)
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

```python
# Univariate Analysis

# 1. Selling Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling Price'], bins=20, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. MRP
plt.figure(figsize=(10, 6))
sns.histplot(df['MRP'], bins=20, kde=True)
plt.title('Distribution of Maximum Retail Price (MRP)')
plt.xlabel('MRP')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Items Bought Last Month
plt.figure(figsize=(10, 6))
sns.countplot(x='Items Bought Last Month', data=df)
plt.title('Count of Items Bought Last Month')
plt.xlabel('Items Bought Last Month')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Numeric Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Numeric_Ratings'], bins=20, kde=True)
plt.title('Distribution of Numeric Ratings')
plt.xlabel('Numeric Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Offer Percentage
plt.figure(figsize=(10, 6))
sns.histplot(df['Offer%'], bins=20, kde=True)
plt.title('Distribution of Offer Percentage')
plt.xlabel('Offer Percentage')
plt.ylabel('Frequency')
plt.show()
```

### Inferneces


* Selling Price:
The distribution of selling prices is right-skewed, indicating that the majority of products have lower selling prices, with a few higher-priced items.

* Maximum Retail Price (MRP):
The distribution of MRPs shows variability, with some products having higher maximum retail prices.

* Items Bought Last Month:
The count plot for items bought last month provides insights into the popularity of different quantities. It appears that a moderate number of products were bought across various quantities.

* Numeric Ratings:
The distribution of numeric ratings shows that a substantial number of products have high ratings, indicating overall positive customer satisfaction.

* Offer Percentage:
The distribution of offer percentages reveals that discounts are prevalent, with a range of discount values across the products.

* Overall Observations:
The dataset contains a mix of products with varying selling prices, MRPs, and offer percentages.
Customer satisfaction, as reflected in numeric ratings, tends to be positively skewed, with a higher concentration of products receiving positive reviews.
The availability of discounts is common, with different products having varied offer percentages.

```python
# Bivariate Analysis

# 1. Selling Price vs. MRP
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MRP', y='Selling Price', data=df)
plt.title('Selling Price vs. Maximum Retail Price (MRP)')
plt.xlabel('MRP')
plt.ylabel('Selling Price')
plt.show()
```

```python
# 2. Items Bought Last Month vs. Numeric Ratings
plt.figure(figsize=(10, 6))
sns.boxplot(x='Items Bought Last Month', y='Numeric_Ratings', data=df)
plt.title('Items Bought Last Month vs. Numeric Ratings')
plt.xlabel('Items Bought Last Month')
plt.ylabel('Numeric Ratings')
plt.show()
```

```python
# 3. Brand vs. Offer Percentage
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='Offer%', data=df)
plt.title('Brand vs. Offer Percentage')
plt.xlabel('Brand')
plt.ylabel('Offer Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Numeric Ratings vs. Offer Percentage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Numeric_Ratings', y='Offer%', data=df)
plt.title('Numeric Ratings vs. Offer Percentage')
plt.xlabel('Numeric Ratings')
plt.ylabel('Offer Percentage')
plt.show()
```

### Inferneces

* Selling Price vs. MRP:
There is a positive correlation between Selling Price and MRP. As the Maximum Retail Price (MRP) increases, the Selling Price tends to increase. This suggests that products with higher MRPs are sold at higher prices.

* Items Bought Last Month vs. Numeric Ratings:
The boxplot reveals no clear trend between the quantity of items bought last month and numeric ratings. However, products with higher quantities purchased tend to have a wider range of ratings. This could indicate that popular products receive a diverse set of reviews.

* Brand vs. Offer Percentage:
Different brands exhibit variations in offer percentages. Some brands consistently offer higher discounts compared to others. This insight is valuable for both customers looking for discounts and brands aiming to optimize pricing strategies.

* Numeric Ratings vs. Offer Percentage:
There is no clear correlation between numeric ratings and offer percentages. This suggests that the level of discount offered does not necessarily influence the overall customer ratings. Products with lower ratings may still have high offer percentages.

* Overall Observations:
Selling Price tends to follow the trend of Maximum Retail Price, indicating a logical pricing structure.
The relationship between Items Bought Last Month and Numeric Ratings is not straightforward, indicating that purchase quantity alone may not strongly influence ratings.
Brands have distinct strategies for offering discounts, with some consistently providing higher discounts.
No clear correlation is observed between Numeric Ratings and Offer Percentage, emphasizing that customers' perceptions are not solely influenced by discounts.
