```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_tablets_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '100'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)


df
```

```python
df.Brand.unique()
```

```python
df.Brand.value_counts()[:40]
```

```python
import random

famous_tab_brands = [
    'Apple (iPad)','Samsung (Galaxy Tab)','Amazon (Kindle Fire)','Lenovo (Tab series)','Huawei (MediaPad)',
    'Microsoft (Surface)','ASUS (ZenPad)','Google (Pixel Slate)','Sony (Xperia Tablet)','Xiaomi (Mi Pad)'
]

# Replace existing values in "Brand" column with random famous women's shoes brands
df['Brand'] = df['Brand'].apply(lambda x: random.choice(famous_tab_brands))
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

```python
df.info()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].astype(int)
```

```python
# Univariate Analysis

# 1. Selling Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling Price'], bins=20, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. MRP
plt.figure(figsize=(10, 6))
sns.histplot(df['MRP'], bins=20, kde=True)
plt.title('Distribution of Maximum Retail Price (MRP)')
plt.xlabel('MRP')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Items Bought Last Month
plt.figure(figsize=(10, 6))
sns.countplot(x='Items Bought Last Month', data=df)
plt.title('Count of Items Bought Last Month')
plt.xlabel('Items Bought Last Month')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Numeric Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Numeric_Ratings'], bins=20, kde=True)
plt.title('Distribution of Numeric Ratings')
plt.xlabel('Numeric Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Offer Percentage
plt.figure(figsize=(10, 6))
sns.histplot(df['Offer%'], bins=20, kde=True)
plt.title('Distribution of Offer Percentage')
plt.xlabel('Offer Percentage')
plt.ylabel('Frequency')
plt.show()
```

### Inferences


Selling Price:

The Selling Price of Galaxy tablets ranges from relatively low to high values, indicating a diverse pricing structure within the Galaxy tablet product line.

MRP (Maximum Retail Price):

The Maximum Retail Price (MRP) varies across the Galaxy tablet models, reflecting differences in features, specifications, and positioning in the market.

Items Bought Last Month:

The quantity of items bought last month ranges across the Galaxy tablet models, indicating variations in popularity and demand. Some tablets have higher recent sales than others.

Numeric Ratings:

Numeric Ratings for Galaxy tablets vary, with ratings ranging from 3.1 to 4.4 out of 5 stars. Customers' perceptions of different tablet models vary in terms of satisfaction and overall experience.

Offer Percentage:

Offer percentages for Galaxy tablets range from single digits to above 80%. This reflects diverse pricing strategies, with some tablets having higher discounts, potentially to stimulate sales.
Overall Univariate Analysis Inferences:

Galaxy tablets have a diverse pricing structure, catering to different market segments.
Variability in Ratings and Total Ratings suggests differences in customer satisfaction and the overall market reception of different tablet models.
Offer percentages vary, indicating distinct pricing and promotional strategies for different tablets within the Galaxy brand.
