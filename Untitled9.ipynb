```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_PowerBank_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
df['Items Bought Last Month'].unique()
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '200'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
```

```python
df
```

```python
df['Brand'].unique()
```

```python
df.Brand.value_counts()[:50]
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

```python
df.info()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].astype(int)
```

```python
# Univariate Analysis

# 1. Selling Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling Price'], bins=20, kde=True)
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. MRP
plt.figure(figsize=(10, 6))
sns.histplot(df['MRP'], bins=20, kde=True)
plt.title('Distribution of Maximum Retail Price (MRP)')
plt.xlabel('MRP')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Items Bought Last Month
plt.figure(figsize=(10, 6))
sns.countplot(x='Items Bought Last Month', data=df)
plt.title('Count of Items Bought Last Month')
plt.xlabel('Items Bought Last Month')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Numeric Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Numeric_Ratings'], bins=20, kde=True)
plt.title('Distribution of Numeric Ratings')
plt.xlabel('Numeric Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Offer Percentage
plt.figure(figsize=(10, 6))
sns.histplot(df['Offer%'], bins=20, kde=True)
plt.title('Distribution of Offer Percentage')
plt.xlabel('Offer Percentage')
plt.ylabel('Frequency')
plt.show()
```

### Inferences

* Selling Price:
The distribution of selling prices is right-skewed, indicating that there are relatively few products with very high selling prices.
Most products fall within a lower to mid-range selling price, with a few outliers having significantly higher prices.

* Maximum Retail Price (MRP):
Similar to the selling price, the distribution of MRP is right-skewed, suggesting a concentration of products with lower MRPs.
There are relatively few products with higher MRPs, and this aligns with the common pricing structure in the market.

* Items Bought Last Month:
The distribution of items bought last month is right-skewed, indicating that a majority of products have lower sales volume.
There are a few products that have significantly higher sales volumes, potentially popular items.

* Offer Percentage:
The distribution of offer percentages is relatively uniform, suggesting that various products have different discount rates.
A significant number of products have offer percentages between 0% and 20%, with fewer products having higher discount rates.

```python
# Bivariate Analysis

# 1. Selling Price vs. MRP
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MRP', y='Selling Price', data=df)
plt.title('Selling Price vs. MRP')
plt.xlabel('MRP')
plt.ylabel('Selling Price')
plt.show()
```

```python
# 3. Brand vs. Offer Percentage
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='Offer%', data=df)
plt.title('Brand vs. Offer Percentage')
plt.xlabel('Brand')
plt.ylabel('Offer Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Numeric Ratings vs. Offer Percentage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Numeric_Ratings', y='Offer%', data=df)
plt.title('Numeric Ratings vs. Offer Percentage')
plt.xlabel('Numeric Ratings')
plt.ylabel('Offer Percentage')
plt.show()
```

### Inferences

Selling Price vs. MRP:

The positive correlation between Selling Price and Maximum Retail Price (MRP) suggests that products with higher MRPs are generally priced higher. This aligns with a logical pricing strategy where the Selling Price tends to reflect the higher perceived value associated with a higher MRP.

Items Bought Last Month vs. Numeric Ratings:

The absence of a clear trend between the quantity of items bought last month and numeric ratings indicates that the popularity of a product, measured by the number of purchases, doesn't necessarily translate into consistent higher or lower ratings. Products with higher purchase quantities exhibit a wider range of ratings, emphasizing the diverse opinions of customers.

Brand vs. Offer Percentage:

The variability in offer percentages across different brands indicates that each brand has a unique approach to pricing and discounting. Some brands consistently offer higher discounts, potentially as part of their marketing or sales strategy, catering to customers looking for discounted products.

Numeric Ratings vs. Offer Percentage:

The lack of a clear correlation between numeric ratings and offer percentages suggests that the level of discount a product carries does not significantly impact customer ratings. This highlights that customers' perceptions and evaluations are influenced by factors beyond just the price, and other product attributes play a crucial role in shaping reviews.

* Overall Observations:
Logical Pricing Structure: The positive correlation between Selling Price and MRP indicates a logical and consistent pricing structure, where the perceived value reflected in the MRP influences the actual selling price.

Diverse Customer Opinions: The lack of a straightforward relationship between purchase quantity and ratings implies that popularity alone does not guarantee consistently positive or negative reviews. Products with higher purchase quantities have a broader spectrum of ratings, emphasizing the diverse opinions of customers.

Brand-Specific Discounting Strategies: Brands exhibit unique discounting strategies, with variations in offer percentages. This insight is valuable for both customers seeking discounts and brands refining their pricing strategies.

Multi-faceted Customer Evaluations: The absence of a strong correlation between numeric ratings and offer percentages underscores that customers consider multiple factors beyond discounts when evaluating and rating products. Product quality, features, and overall satisfaction play significant roles in shaping customer reviews.
