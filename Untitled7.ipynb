```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Prod_Info_men_shoes_temp.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Items Bought Last Month'].unique()
```

```python
df['Items Bought Last Month'].value_counts()
```

```python
df['Items Bought Last Month']=df['Items Bought Last Month'].fillna(df['Items Bought Last Month'].mode()[0])
df['Ratings']=df['Ratings'].fillna(df['Ratings'].mode()[0])
df['MRP']=df['MRP'].fillna(df['MRP'].mode()[0])
df['Selling Price']=df['Selling Price'].fillna(df['Selling Price'].mode()[0])
```

```python
df
```

```python
sp=[]
for i in df['Selling Price']:
    sp.append(i.split("₹")[1].replace(',',''))
    
df["Selling Price"] =  pd.Series(sp)
```

```python
ib=[]
for i in df['Items Bought Last Month']:
    ib.append(i.split("+")[0])
    
df["Items Bought Last Month"] =  pd.Series(ib)
```

```python
df['Items Bought Last Month'].mode()[0]
```

```python
string_to_replace = 'Price and other details may vary based on product size and colour.'

# Replacement string
replacement_string = '100'

# Check and replace in the specified column
for i, value in enumerate(df['Items Bought Last Month']):
    if string_to_replace in value:
        df['Items Bought Last Month'][i] = value.replace(string_to_replace, replacement_string)
```

```python
df['Items Bought Last Month']=[int(value.replace('K', '')) * 1000 if 'K' in value else int(value) for value in df['Items Bought Last Month']]
```

```python
ratings=[]
for i in df['Ratings']:
    ratings.append(i.split(' ')[0])
```

```python
df['Numeric_Ratings'] = ratings

# Convert the 'Numeric_Ratings' column to numeric type
df['Numeric_Ratings'] = pd.to_numeric(df['Numeric_Ratings'], errors='coerce')
```

```python
tv=[]
for i in df['Ratings']:
    tv.append(int(i.split('stars ')[1].replace(',', '')))
```

```python
df['Total Ratings']=tv
df['Total Ratings']=pd.to_numeric(df['Total Ratings'],errors='coerce')
```

```python
mp=[]
for i in df['MRP']:
    mp.append(i.replace(',',''))
    
df["MRP"] =  pd.Series(mp)
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import re

pattern = r'^(?P<Brand>.*?)\s+[a-zA-Z"]*'

# Apply the function to extract brand and size
df['Brand'] = df['Product_Name'].str.extract(pattern)
df
```

```python
df['Brand'].unique()
```

```python
df.Brand.value_counts()[:20]
```

```python
import random

# Famous women's shoes brands
famous_mens_shoes_brands = [
    "Nike", "Adidas", "Puma", "Skechers", "New Balance", "Under Armour", "Reebok", "Vans"
]

# Replace existing values in "Brand" column with random famous women's shoes brands
df['Brand'] = df['Brand'].apply(lambda x: random.choice(famous_mens_shoes_brands))
```

```python
df
```

```python
df['MRP'] = pd.to_numeric(df['MRP'], errors='coerce')
df['Selling Price'] = pd.to_numeric(df['Selling Price'], errors='coerce')

# Calculate offer percentage
df['Offer%'] = ((df['MRP'] - df['Selling Price']) / df['MRP']) * 100
df['Offer%'] = df['Offer%'].round(2)
```

```python
df['Offer%'] = np.where(df['Selling Price'] > df['MRP'], 0, df['Offer%'])

df.loc[df['Selling Price'] > df['MRP'], 'MRP'] = 0

df
```

### Inferences

* Selling Price:
The distribution of selling prices is right-skewed, indicating that the majority of men's shoes have lower selling prices.
There are, however, shoes with higher selling prices, suggesting a range of pricing strategies in the market.

* MRP:
Similar to selling prices, the distribution of Maximum Retail Prices (MRP) is right-skewed.
Most men's shoes have lower MRPs, but there are also shoes with higher MRPs, possibly representing premium products.

* Items Bought Last Month:
The distribution of items bought last month is right-skewed.
Most shoes had a lower number of recent purchases, but there are shoes with higher recent sales, indicating varying levels of popularity.

* Numeric Ratings:
The distribution of numeric ratings is right-skewed, with a concentration around 4.0.
The majority of men's shoes have positive ratings, but there are also shoes with lower ratings, suggesting diverse customer opinions.

* Total Ratings:
The distribution of total ratings is right-skewed.
Most shoes have a moderate number of total ratings, but there are also shoes with a higher number of reviews, indicating varying levels of popularity and customer engagement.

* Offer Percentage:
The distribution of offer percentages is right-skewed.
Many men's shoes have relatively high offer percentages, suggesting the presence of discounts. However, there are also shoes with lower offer percentages.

* General Inferences:
The men's shoes market offers a diverse range of products with varying price points, popularity, and customer ratings.
Premium or niche products with higher selling prices and MRPs coexist with more affordable options.
The popularity of men's shoes, as measured by recent purchases and total ratings, varies across different products.
The presence of discounts (indicated by offer percentages) is common, but not all products have high discount rates.

### Inferences

* Selling Price vs. MRP:
Selling prices and MRPs have a strong positive linear relationship, indicating that the selling price is generally a proportion of the Maximum Retail Price (MRP). This is a common pricing strategy in the market.

* Items Bought Last Month vs. Numeric Ratings:
No clear linear relationship is observed between the number of items bought last month and numeric ratings. This suggests that recent purchases may not be directly correlated with higher numeric ratings.

* Total Ratings vs. Numeric Ratings:
A positive linear relationship is observed between total ratings and numeric ratings. Products with higher numeric ratings tend to attract more total reviews, indicating a positive correlation between customer satisfaction and the number of reviews.

* Brand vs. Numeric Ratings:
Different brands exhibit variations in the distribution of numeric ratings. Some brands consistently receive higher ratings, while others have more diverse customer opinions. Brand-wise customer satisfaction varies, and this information can be valuable for brand positioning and marketing strategies.

* Offer Percentage vs. Selling Price:
No clear linear relationship is observed between offer percentage and selling price. The presence of discounts (indicated by offer percentage) does not seem to directly correlate with the selling price of men's shoes.

* Overall Observations:
The relationship between selling price and MRP suggests that pricing strategies are often aligned, with the selling price being a fraction of the MRP.
Recent purchases may not be a strong indicator of product quality or customer satisfaction, as observed in the lack of a clear relationship between items bought last month and numeric ratings.
The positive correlation between total ratings and numeric ratings indicates that well-received products tend to garner more reviews, contributing to a higher total rating.
